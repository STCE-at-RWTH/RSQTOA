```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df_results_ann = pd.read_csv('./ann/results.csv')
df_results_grid = pd.read_csv('./grid/results.csv')
df_results_interpolator = pd.read_csv('./interpolator/results.csv')
df_results_lr = pd.read_csv('./elasticnet/results.csv')
```

```python
df_time = pd.concat([df_results_lr['train_time'], df_results_ann['train_time'], df_results_grid['train_time'], df_results_interpolator['train_time']], axis=1, keys=['ElasticNet', 'ANN', 'RSQTOA + Grid', 'RSQTOA + Interpolator'])
df_time['fold'] = list(range(1, 21))
df_time
```

```python
df_time.plot(x='fold', y=['ANN', 'ElasticNet', 'RSQTOA + Grid', 'RSQTOA + Interpolator'], marker='o')
plt.xticks(np.arange(1, 21, 2.0))
plt.yticks(np.arange(0, 20, 2.0))
plt.ylabel('Training Time (sec)')
plt.xlabel('Fold')
plt.savefig('synthetic_train_time_comparison.png', dpi = 300)
```

```python
df_rmse = pd.concat([df_results_lr['test_rmse'], df_results_ann['test_rmse'], df_results_grid['test_rmse'], df_results_interpolator['test_rmse']], axis=1, keys=['ElasticNet', 'ANN', 'RSQTOA + Grid', 'RSQTOA + Interpolator'])
df_rmse['fold'] = list(range(1, 21))
df_rmse
```

```python
df_rmse.plot(x='fold', y=['ANN', 'ElasticNet', 'RSQTOA + Grid', 'RSQTOA + Interpolator'], marker='o')
plt.xticks(np.arange(1, 21, 1.0))
plt.yticks(np.arange(0, 0.07, 0.007))
plt.ylabel('RMSE')
plt.xlabel('Fold')
plt.show()
```

```python
df_r2 = pd.concat([df_results_lr['test_r2'], df_results_ann['test_r2'], df_results_grid['test_r2'], df_results_interpolator['test_r2']], axis=1, keys=['ElasticNet', 'ANN', 'RSQTOA + Grid', 'RSQTOA + Interpolator'])
df_r2['fold'] = list(range(1, 21))
df_r2
```

```python
df_r2.plot(x='fold', y=['ANN', 'ElasticNet', 'RSQTOA + Grid', 'RSQTOA + Interpolator'], marker='o')
plt.xticks(np.arange(1, 21, 2.0))
plt.yticks(np.arange(0.9, 1.1, 0.05))
plt.ylabel('$R^2$ Score')
plt.xlabel('Fold')
plt.savefig('synthetic_r2_comparison.png', dpi = 300)
```

```python
df_loss = pd.DataFrame(columns=['ann_train_loss', 'ann_val_loss'])
for i in range(1, 21):
    df_loss_ann = pd.read_csv('./ann/loss_hist_{}.csv'.format(i))
    df_loss_grid = pd.read_csv('./grid/loss_hist_{}.csv'.format(i))
    df_loss_interpolator = pd.read_csv('./interpolator/loss_hist_{}.csv'.format(i))

    if df_loss.shape[0] is 0:
        df_loss['ann_train_loss'] = df_loss_ann['train_loss']
        df_loss['ann_val_loss'] = df_loss_ann['val_loss']

        df_loss['grid_train_loss'] = df_loss_grid['train_loss']
        df_loss['grid_val_loss'] = df_loss_grid['val_loss']

        df_loss['interpolator_train_loss'] = df_loss_interpolator['train_loss']
        df_loss['interpolator_val_loss'] = df_loss_interpolator['val_loss']
    else:
        df_loss['ann_train_loss'] += df_loss_ann['train_loss']
        df_loss['ann_val_loss'] += df_loss_ann['val_loss']

        df_loss['grid_train_loss'] += df_loss_grid['train_loss']
        df_loss['grid_val_loss'] += df_loss_grid['val_loss']

        df_loss['interpolator_train_loss'] += df_loss_interpolator['train_loss']
        df_loss['interpolator_val_loss'] += df_loss_interpolator['val_loss']

df_loss /= 20
df_loss
```

```python
plt.plot(df_loss['ann_train_loss'])
plt.plot(df_loss['ann_val_loss'])
plt.yticks(np.arange(0.0, 0.4, 0.08))
# plt.title('Model loss ANN')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='upper right')
plt.savefig('synthetic_residual_ann.png', dpi = 300)
```

```python
plt.plot(df_loss['grid_train_loss'])
plt.plot(df_loss['grid_val_loss'])
plt.yticks(np.arange(0.0, 0.4, 0.08))
# plt.title('Model loss Grid')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='upper right')
plt.savefig('synthetic_residual_grid.png', dpi = 300)
```

```python
plt.plot(df_loss['interpolator_train_loss'])
plt.plot(df_loss['interpolator_val_loss'])
plt.yticks(np.arange(0.0, 0.4, 0.08))
# plt.title('Model loss Interpolator')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='upper right')
plt.savefig('synthetic_residual_interpolator.png', dpi = 300)
```
